```python
def glc(semente,multiplicador,modulo,tamanho):
  U = np.zeros(tamanho)
  n = np.zeros(tamanho)
  n[0] = (semente * multiplicador) % modulo
  U[0] = n[0] / modulo

  for i in range(1,tamanho):
    n[i] = (n[i-1] * multiplicador) % modulo
    U[i] = n[i] / modulo

  # print(n)
  return U
```
